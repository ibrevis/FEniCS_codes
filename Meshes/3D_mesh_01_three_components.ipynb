```python
import fenics as fn
import numpy as np
#
import mshr
from math import pi, sin, cos, sqrt
```

```python
#from fenics import *


# Parameters
R = 1.1
r = 0.4
t = 10
x = R*cos(float(t) / 180 * pi)
y = 0
z = R*sin(t)

# Create geometry
s1 = mshr.Sphere(fn.Point(0, 0, 0), 1)
s2 = mshr.Sphere(fn.Point(x, y, z), r)
b1 = mshr.Box(fn.Point(-2, -2, -0.03), fn.Point(2, 2, 0.03))
geometry = s1 - s2 - b1

# Create mesh
mesh = mshr.generate_mesh(geometry, 32)
```
